# 預處理
## 環境準備

```python
%matplotlib inline
import cv2
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

## 用OpenCV找出臉部的位置

```python
input_image = cv2.imread('./data/example/LIU.jpg',cv2.IMREAD_COLOR) #将档读入成彩色的
face_cascade = cv2.CascadeClassifier("./model/haarcascades/haarcascade_frontalface_default.xml")
face_img = input_image.copy() #建立复本,避免影响原本的图像
face_img_RGB = cv2.cvtColor(face_img, cv2.COLOR_BGR2RGB) # OpenCV用BGR顺序,而matplotlib用RGB, 所以必须做一次转换 
face_rects = face_cascade.detectMultiScale(face_img_RGB)  #取得脸的位置
for (x,y,w,h) in face_rects:  #将画面中所有的脸画出来
    cv2.rectangle(face_img_RGB, (x,y), (x+w,y+h), (255,0,0),  3) 
plt.imshow(face_img_RGB)
plt.show()
```
